```python
raw_email = """\
From: John Doe <john.doe@enron.com>
To: Jane Smith <jane.smith@enron.com>, boss@enron.com
Cc: "Support Team" <support@enron.com>
Subject: Meeting update
Date: Mon, 12 Nov 2001 09:15:30 -0800
Message-ID: <12345@enron.com>
Content-Type: multipart/alternative; boundary="ABC123"

--ABC123
Content-Type: text/plain; charset="utf-8"

Hi team,

Just confirming today's meeting at 2 PM.

Best,
John

--ABC123
Content-Type: text/html; charset="utf-8"

<html><body><p>Hi team,</p><p>Just confirming today's meeting at 2 PM.</p><p>Best,<br>John</p></body></html>

--ABC123--
"""
```

```python
from email import message_from_string

raw_email = message_from_string(raw_email)
```

```python
from email_parser import parser

parsed = parser.parse_email(raw_email)
```
